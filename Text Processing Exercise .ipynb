```python
# imported packages

import pandas as pd
import string
import re

#NLTK packages 
import nltk
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
df = pd.read_csv("AmazonReviews.csv")
df
```

```python
# selecting the Column Text to clean separetly into a new var
TextCleaning = df['Text']
TextCleaning
```

```python
# check for the number of duplicates
DuplicatesRemoved = TextCleaning.duplicated()
print(DuplicatesRemoved.sum())
```

```python
#remove duplicated values
DuplicatesRemoved = TextCleaning.drop_duplicates(keep = 'first')
DuplicatesRemoved
```

```python
#Remove HTMl tags

def get_cleaned_textdata(text):
    #Remove HTMl tags
    cleanedText = re.sub(r'<.*?>','', text)
    #replace punctuation and Special Characters
    cleanedText = ''.join([i if i not in string.punctuation else '' for i in cleanedText])
    #remove digits
    cleanedText = re.sub("[\d+]",'',text)
    #replace multipable spcapes with single space
    cleanedText = re.sub(r'\s+','', text)
    #convert to lower case letters
    cleanedText = cleanedText.lower()
    return cleanedText
```

```python
fullCleaned = DuplicatesRemoved.apply(get_cleaned_textdata)
```

```python
def remove_stopwords(text):
    wordClear = []
    for word in text:
        if word not in stopwords:
            wordClear.append(word)
    return wordClear
```
