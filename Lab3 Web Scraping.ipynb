```python
#imported packages 

import requests
from bs4 import BeautifulSoup
import csv
```

```python
#web to scrap wikipedia page with list of largest companies in malaysia by revenue
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_Malaysia'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')[1]
headers = table.find_all('th')
```

```python
print(headers)
```

```python
headers_table = [title.text.strip() for title in headers]
print(headers_table)
```

## import data into a dataframe

```python
import pandas as pd 
```

```python
df = pd.DataFrame(columns = headers_table)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
